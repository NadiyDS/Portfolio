# **Titanic**
**Introduction**\
The sinking of the Titanic is one of the most famous maritime disasters in history. On the night of April 14–15, 1912, the ship struck an iceberg during its maiden voyage and sank. Of the more than 2,200 passengers and crew on board, only about 31% survived.\
This project focuses on predicting the survival of Titanic passengers based on their characteristics. The data is provided by Kaggle as part of the classic competition “Titanic: Machine Learning from Disaster.”\
The goal of this study is to predict whether a passenger survived the Titanic disaster using a trained machine learning model.

**Datasets**\
train.csv — the training set containing passenger information along with the target label (Survived).\
test.csv — the test set containing passenger information without the target label. The task is to predict the survival of these passengers.

## **Feature Engineering**
After performing the exploratory data analysis (EDA), it became clear that the original features do not always fully capture important information about the passengers. To improve model performance, it is necessary to transform the existing data and create new informative features.

```python
# Copying the data
train = df_train.copy()
test = df_test.copy()

def create_features(df):
    df = df.copy()
    
    # Age - Filling missing values using the median based on passenger class and sex
    df["Age"] = df.groupby(["Pclass","Sex"])["Age"].transform(lambda x: x.fillna(x.median()))
    
    # AgeGroup - We group the Age feature into categories: Child, Teenager, Youth, Adult, and Senior.
    df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,80], labels=['Child','Teen','YoungAdult','Adult','Senior'])
    
    # Deck – extracting the deck number from the cabin number
    df["Deck"] = df["Cabin"].str[0].fillna("Unknown")
    
    # FamilySize – total number of family members on board (SibSp + Parch + 1)
    df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
    df['FamilySizeCat'] = pd.cut(df['FamilySize'], bins=[0,1,4,11], labels=['Alone','Small','Large'])
    
    # IsAlone – indicator of whether the passenger was traveling alone
    df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
        
    # Title – extract the passenger’s title from their name (e.g., Mr, Mrs, Miss, etc.)
    df["Title"] = df["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
    df["Title"] = df["Title"].replace(['Lady','Countess','Capt','Col','Don','Dr','Major',
                                       'Rev','Sir','Jonkheer','Dona'], 'Rare')
    df["Title"] = df["Title"].map({"Master":0, "Miss":1, "Ms":1, "Mme":1, "Mlle":1, 
                                   "Mr":2, "Mrs":1, "Rare":3}).fillna(3)
    
    # Fill missing values in Fare with the median based on Pclass and Embarked.
    df['Fare'] = df.groupby(['Pclass','Embarked'])['Fare'].transform(lambda x: x.fillna(x.median()))
    df['Fare'] = df['Fare'].fillna(df['Fare'].median())
    
    # Fare_log
    df["Fare_log"] = np.log1p(df["Fare"])

    # For Embarked, fill missing values with the mode.
    if 'Embarked' in df.columns:
        df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
        
    # Drop unnecessary columns
    df = df.drop(columns=["Cabin", "Name", "Ticket"], errors='ignore')
    
    return df

train = create_features(train)
test = create_features(test)
train.head()
```

```python
# === Plot graphs for the new features ===
new_cat_cols = ["AgeGroup", "Deck", "IsAlone", "Title", "FamilySizeCat"]

for col in new_cat_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12,4))
    
    # Feature distribution
    sns.countplot(x=col, data=train, ax=axes[0])
    axes[0].set_title(f"Distribution of {col}")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Count")
    axes[0].tick_params(axis='x', rotation=30)
    
    # Distribution by Survived
    sns.countplot(x=col, hue="Survived", data=train, ax=axes[1])
    axes[1].set_title(f"{col} vs Survived")
    axes[1].set_xlabel(col)
    axes[1].set_ylabel("Count")
    axes[1].tick_params(axis='x', rotation=30)
    
    plt.tight_layout()
    plt.show()
```

**Conclusions from the Feature Engineering section:**\
1.New features have been created that can potentially improve model performance:\
    - FamilySize (number of family members on board) — captures social connections of the passenger.\
    - IsAlone — reflects the effect of traveling alone.\
    - Title — highlights social status and may correlate with survival.\
    - Deck (extracted from Cabin) — indirectly indicates the location of the cabin on the ship.\
    - Fare_log (logarithm of Fare) — reduces the impact of outliers and makes the distribution more normal.\
2. Features have been transformed into more informative forms:\
    - Categorical values have been grouped and encoded.\
    - Fare has been log-transformed to mitigate the effect of outliers.\
3. Missing values in Cabin have been treated as a separate category “Unknown”.\
  The goal of this stage has been achieved: the data is now more structured and includes additional informative features that can enhance model training.

## **Data Preprocessing**

The main tasks of preprocessing are: converting categorical features to numerical (encoding), scaling numerical features (if needed for certain algorithms), removing or transforming irrelevant features (e.g., PassengerId, Name, Ticket in their original form), proper preprocessing ensures a clean and structured dataset, allowing machine learning models to identify patterns effectively and make more accurate predictions.

```python
# Encoding Categorical Features
cat_cols = ["Sex","Pclass","Embarked","Deck","AgeGroup","Title","IsAlone","FamilySizeCat"]
combined = pd.concat([train.drop(columns=["Survived"]), test], axis=0)
combined = pd.get_dummies(combined, columns=cat_cols, drop_first=True)

X_train = combined.iloc[:train.shape[0], :]
X_test = combined.iloc[train.shape[0]:, :]
y_train = train["Survived"]

# Scaling Numerical Features
num_cols = ["Age","SibSp","Parch","Fare_log"]
scaler = StandardScaler()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
# Dropping Irrelevant Columns
#train= train.drop(columns=["Cabin", "Name", "Ticket"], errors='ignore')
train= train.drop(columns=["PassangerId"], errors='ignore')
```

## Modeling
The goal of modeling is to select algorithms and configurations that best capture patterns in the data and provide high prediction accuracy.

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import pandas as pd

models = {
    "Logistic Regression": LogisticRegression(max_iter=3000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(probability=True, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, learning_rate=0.05, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=200, random_state=42),
    "XGBoost": XGBClassifier(eval_metric="logloss", random_state=42),
}

# Cross-Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    # Model Evaluation with Cross-Validation
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring="accuracy")
    print(f"{name}: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    # Train the model on the entire train dataset and make predictions on the test dataset.
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    
    # Save the submission file for each model.
    submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": predictions
    })
    
    file_name = f"submission_{name.replace(' ', '_')}.csv"
    submission.to_csv(file_name, index=False)
    print(f"✅ Файл {file_name} сохранен")
```

## **Conclusions:**
- The best models in terms of cross-validation accuracy were Gradient Boosting, Random Forest, and Logistic Regression.
- When submitting predictions to Kaggle, Gradient Boosting achieved the highest accuracy of 0.77272, confirming that this model generalizes best to new data.
- General observations: ensemble models (Random Forest, Gradient Boosting, AdaBoost) performed best because they capture complex relationships between features and reduce overfitting. Logistic Regression showed stable cross-validation accuracy and is suitable as a baseline model, especially after feature scaling.
